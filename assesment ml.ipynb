Concept Based Questions:


1.Multicollinearity is when independent variables in a regression model are highly correlated. It impacts the model by causing unreliable coefficient estimates, reduced interpretability, inflated standard errors, and difficulties in model selection. Techniques to handle multicollinearity include using variance inflation factor (VIF) to identify highly correlated variables, removing or combining correlated variables, and regularization methods like ridge regression or lasso regression.

2.Outliers can significantly impact the regression model by distorting the relationship between variables and leading to inaccurate parameter estimates and reduced model performance.
Techniques to detect and handle outliers include visual inspection of scatter plots, box plots, or leverage plots, statistical methods such as the z-score or modified z-score, and robust regression techniques like Huber regression or Theil-Sen regression to minimize the influence of outliers.2.

3.Overfitting in regression modeling occurs when a model fits the training data too closely and fails to generalize well to new, unseen data. It is problematic because it leads to poor performance and unreliable predictions on new data.

Regularization techniques like Ridge and Lasso regression help address overfitting by adding a penalty term to the regression objective function.

Ridge regression controls overfitting by shrinking the coefficients towards zero, reducing their magnitude but not necessarily eliminating them.
Lasso regression, in addition to shrinking coefficients, can also perform feature selection by driving some coefficients to exactly zero.
Both techniques strike a balance between model complexity and data fit, preventing overfitting and improving model performance.

4.When dealing with non-linear relationships between predictors and the target variable, regression models can be enhanced using polynomial regression.

Polynomial regression captures non-linear patterns by including polynomial terms of the predictors in the regression equation. It allows the model to fit curves or higher-order relationships rather than assuming a linear relationship.

To select an appropriate degree for polynomial regression, one can start with a simple linear regression and gradually increase the degree. The degree represents the highest power of the predictor variable in the equation. The appropriate degree is determined by assessing the model's fit using evaluation metrics (e.g., R-squared, mean squared error) on a validation set. Select the degree that provides the best trade-off between model complexity and performance, avoiding overfitting by not selecting a degree that is too high.
